```python
# df.append()

# Using df.append() to add the 'dec' DataFrame to 'nov'. 
# This method is deprecated and will be removed in future versions of Pandas. 
# It is recommended to use pd.concat instead to avoid FutureWarning.
nov.append(dec, ignore_index=True)  # Warning: Use pd.concat instead.
```

```python
# pd.concat

# To resolve the FutureWarning caused by the deprecated df.append() method, 
# I used pd.concat to concatenate the 'nov' and 'dec' DataFrames. 
# The ignore_index=True argument ensures the index is reset in the combined DataFrame 'regs'.
regs = pd.concat([nov, dec], ignore_index=True)  # Recommended solution.
regs
```

```python
# mullitindex DataFrame

# Creating a MultiIndex DataFrame using pd.concat with keys 'Nov' and 'Dec' 
# to differentiate the rows from the 'nov' and 'dec' DataFrames.
multi = pd.concat([nov, dec], keys=['Nov', 'Dec'])
multi
```

```python
# Concatenating the 'nov' and 'dec' DataFrames along the columns (axis=1).
# This combines the DataFrames side by side, resulting in a new DataFrame.
combined = pd.concat([nov, dec], axis=1)
combined
```

```python
# MultiIndex -> fetch using iloc

# Using .loc to fetch data from the MultiIndex DataFrame 'multi'.
# The DataFrame 'multi' has a MultiIndex created with keys ['Nov', 'Dec'].
# To retrieve data for December, we specify the key 'Dec' along with the row index (4).
# This accesses the specific row in the 'Dec' section of the DataFrame.
# The resulting output will include all columns for the specified row under 'Dec'.
dec_data = multi.loc[('Dec', 4)]
dec_data
```

```python
# inner join

# Performing an inner join between the 'students' DataFrame and the 'regs' DataFrame.
# The merge is based on the 'student_id' column, which is common to both DataFrames.
# The 'how' parameter is set to 'inner', meaning that only rows with matching 'student_id'
# values in both DataFrames will be included in the result. 
# This is useful for filtering out students that have registration records.
result = students.merge(regs, how='inner', on='student_id')
result
```

```python
# left join

# Performing a left join between the 'courses' DataFrame and the 'regs' DataFrame.
# The merge is based on the 'course_id' column, which is common to both DataFrames.
# The 'how' parameter is set to 'left', meaning that all rows from the 'courses' DataFrame
# will be included in the result, along with matching rows from the 'regs' DataFrame.
# If there is no match for a course in 'regs', the resulting DataFrame will have NaN 
# values for the columns from 'regs'. This is useful for retaining all course information,
# even if some courses do not have corresponding registration records.
result = courses.merge(regs, how='left', on='course_id')
result
```

```python
# right join

# Creating a temporary DataFrame 'temp_df' to simulate additional student data.
# This DataFrame contains 'student_id', 'name', and 'partner' columns.
temp_df = pd.DataFrame({
    'student_id': [26, 27, 28],
    'name': ['Nitish', 'Ankit', 'Rahul'],
    'partner': [28, 26, 17]
})

# Concatenating 'temp_df' with the existing 'students' DataFrame to include the new student records.
students = pd.concat([students, temp_df], ignore_index=True)

# Performing a right join between the updated 'students' DataFrame and the 'regs' DataFrame.
# The merge is based on the 'student_id' column, which is common to both DataFrames.
# The 'how' parameter is set to 'right', meaning that all rows from the 'regs' DataFrame
# will be included in the result, along with matching rows from the 'students' DataFrame.
# If there is no match for a student in 'students', the resulting DataFrame will have NaN 
# values for the columns from 'students'. This is useful for retaining all registration records,
# even if some registrations do not correspond to students in the 'students' DataFrame.
result = students.merge(regs, how='right', on='student_id')
result
```

```python
# left join


# Note: If we place the 'regs' DataFrame first and the 'students' DataFrame second (in parentheses) 
# while setting the 'how' parameter to 'left', it will perform the same operation as described above.
# This means that all rows from 'regs' will be included in the result, 
# along with matching rows from 'students', with NaN values where there are no matches.
regs.merge(students, how='left', on='student_id')
```

```python
# Outer Join

# Performing an outer join between the 'students' DataFrame and the 'regs' DataFrame.
# The merge is based on the 'student_id' column, which is common to both DataFrames.
# The 'how' parameter is set to 'outer', meaning that all rows from both 'students' 
# and 'regs' will be included in the result. Where there are no matches, 
# the resulting DataFrame will have NaN values for the columns from the DataFrame 
# that does not have a corresponding match.
# The .tail(10) method is used to display the last 10 rows of the resulting DataFrame.
students.merge(regs, how='outer', on='student_id').tail(10)
```

```python
# 1. Find Total Revenue Generated

# Merging the 'regs' DataFrame with the 'courses' DataFrame based on the 'course_id' column.
# The 'how' parameter is set to 'inner', meaning that only rows with matching 'course_id' 
# values in both DataFrames will be included in the result. 
# We then access the 'price' column and calculate the total revenue generated 
# by summing all the prices.
total = regs.merge(courses, how='inner', on='course_id')['price'].sum()
total
```

```python
# 2. Find Month-by-Month Revenue

# Concatenating the 'nov' and 'dec' DataFrames with keys 'Nov' and 'Dec' 
# to differentiate between the two months. The resulting DataFrame is reset to have a sequential index.
temp_df = pd.concat([nov, dec], keys=['Nov', 'Dec']).reset_index()

# Merging the concatenated DataFrame 'temp_df' with the 'courses' DataFrame 
# based on the 'course_id' column to bring in the 'price' information.
# We then group the merged DataFrame by 'level_0' (which represents the month) 
# and calculate the total revenue for each month by summing the 'price' values.
month_revenue = temp_df.merge(courses, on='course_id').groupby('level_0')['price'].sum()
month_revenue
```

```python
# 3. Print the Registration Table

# Merging the 'regs' DataFrame with the 'students' DataFrame on 'student_id' 
# to get student names and then merging with the 'courses' DataFrame on 'course_id' 
# to retrieve course names and prices. 
# Finally, we select only the relevant columns: 'name', 'course_name', and 'price'.
registration_table = regs.merge(students, on='student_id').merge(courses, on='course_id')[['name', 'course_name', 'price']]
registration_table
```

```python
# 4. Plot Bar Chart for Revenue per Course

# Merging the 'regs' DataFrame with the 'courses' DataFrame on 'course_id' 
# to get the course names and their corresponding prices. 
# The merged DataFrame is then grouped by 'course_name', 
# and the total price (revenue) is summed for each course.
# Finally, a bar chart is plotted to visualize the total revenue generated for each course.
regs.merge(courses, on='course_id').groupby('course_name')['price'].sum().plot(kind='bar', 
    title='Total Revenue per Course', 
    xlabel='Course Name', 
    ylabel='Total Revenue')
```

```python
# 5. Find Students Who Enrolled in Both Months

# Using numpy's intersect1d function to identify student IDs that are present 
# in both the November ('nov') and December ('dec') registration DataFrames. 
# This will give us the list of student IDs who enrolled in both months.
enrolled_both_months = np.intersect1d(nov['student_id'], dec['student_id'])
enrolled_both_months
```

```python
# Solving the above problem 4. using join

# Merging the 'nov' and 'dec' DataFrames on 'student_id' to find students 
# who registered in both November and December. The resulting DataFrame 
# is filtered to select only the 'student_id' column, sorted in ascending 
# order, and unique student IDs are extracted.
nov.merge(dec, on='student_id')['student_id'].sort_values().unique()
```

```python
# Filtering the 'students' DataFrame to find students who enrolled in both 
# November and December. This is achieved by checking if 'student_id' 
# is in the array of common student IDs obtained earlier.
students[students['student_id'].isin(enrolled_both_months)]
```

```python
# 6. find course that got no enrollment

# Finding courses that received no enrollment by comparing the 'course_id' 
# from the 'courses' DataFrame with the 'course_id' in the 'regs' DataFrame. 
# The np.setdiff1d function identifies course IDs present in 'courses' 
# but not in 'regs', and the resulting list is used to filter the 'courses' DataFrame.
no_enrollment_courses = courses[courses['course_id'].isin(np.setdiff1d(courses['course_id'], regs['course_id']))]
no_enrollment_courses 
```

```python
# Solving the above problem 6. using join

# Finding courses that did not have any enrollment in either November or December.
# This is done by merging the 'nov' and 'dec' DataFrames on 'course_id' 
# using an outer join, then extracting unique course IDs from the merged result.
# The '~' operator is used to filter the 'courses' DataFrame to return courses 
# whose 'course_id' is not present in the combined enrollment list.
no_enrollment_courses_nov_dec = courses[~courses['course_id'].isin(nov.merge(dec, how='outer', on='course_id')['course_id'].unique())]
no_enrollment_courses_nov_dec
```

```python
# 7. Find students who did not enroll in any courses

# Create a list of student IDs that are present in the 'students' DataFrame
# but not in the 'regs' DataFrame, indicating these students did not enroll in any courses.
student_id_list = np.setdiff1d(students['student_id'], regs['student_id'])

# Filter the 'students' DataFrame to include only those students whose IDs are in 'student_id_list'
# Then, get the count of these students by checking the number of rows in the resulting DataFrame.
students[students['student_id'].isin(student_id_list)].shape[0]
```

```python
# Solving the above problem 7 with another method

# Use the 'isin' method to create a boolean mask that identifies which student IDs in the 'students' DataFrame
# are not present in the 'regs' DataFrame. The '~' operator negates the boolean values.
# The 'unique()' method is used to ensure we are checking against unique student IDs from 'regs'.
# Finally, the 'sum()' function counts the number of True values, indicating the number of students
# who did not enroll in any courses.
(~students['student_id'].isin(regs['student_id'].unique())).sum()
```

```python
# 8. Print student name and partner name for all enrolled students

# Perform a self join on the 'students' DataFrame to match students with their partners.
# The 'left_on' parameter specifies the 'partner' column from the left DataFrame (students),
# while the 'right_on' parameter specifies the 'student_id' column from the right DataFrame (also students).
# The result will contain pairs of names, with 'name_x' representing the student's name
# and 'name_y' representing the partner's name.
students.merge(students, how='inner', left_on='partner', right_on='student_id')[['name_x', 'name_y']]
```

```python
# 9. Find the top 3 students with the most enrollments

# Merge the 'regs' DataFrame with the 'students' DataFrame on 'student_id' 
# to associate students with their enrollment records.
# Group the merged DataFrame by 'student_id' and 'name', and count the number of enrollments 
# (which is simply the number of rows for each student).
# Sort the results in descending order to get the students with the highest enrollment counts,
# and retrieve the top 3 students.
regs.merge(students, on='student_id').groupby(['student_id', 'name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 10. Find the top 3 students who spent the most amount of money on courses.

# First, merge the 'regs' DataFrame with the 'students' DataFrame on 'student_id' 
# to link each enrollment with the corresponding student.
# Then, merge the result with the 'courses' DataFrame on 'course_id' 
# to access the price of each course.
# Group the merged DataFrame by 'student_id' and 'name', and calculate the total amount spent 
# by summing the 'price' for each student.
# Sort the results in descending order to identify the students who spent the most,
# and retrieve the top 3 students.
regs.merge(students, on='student_id').merge(courses, on='course_id').groupby(['student_id', 'name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
# Alternate syntax for performing an inner join

# The 'pd.merge()' function is used instead of 'DataFrame.merge()'.
# This explicitly calls the 'pandas' merge function and performs an inner join 
# between 'students' and 'regs' DataFrames based on the common 'student_id' column.
pd.merge(students, regs, how='inner', on='student_id')
```

```python
# 1.find top 3 studiums with highest sixes/match ratio

# Merge the 'delivery' DataFrame with the 'matches' DataFrame.
# This combines information based on 'match_id' from 'delivery' and 'id' from 'matches'.
temp_df = delivery.merge(matches, left_on='match_id', right_on='id')

# Filter the merged DataFrame to include only the rows where the 'batsman_runs' equals 6,
# indicating that a six was hit.
six_df = temp_df[temp_df['batsman_runs'] == 6]

# Group the filtered DataFrame by 'venue' (stadium) and count the number of sixes 
# hit at each venue.
num_sixes = six_df.groupby('venue')['venue'].count()

# Count the total number of matches held at each venue.
num_matches = matches['venue'].value_counts()

# Calculate the ratio of sixes hit to the number of matches played at each venue,
# sort the result in descending order, and get the top 3 stadiums.
(num_sixes / num_matches).sort_values(ascending=False).head(3)
```

```python
# 2.find orange cap holder of all the seasons

# Group the DataFrame by 'season' and 'batsman', then sum the 'batsman_runs'
# Reset the index to convert the grouped data back to a DataFrame
# Sort the values by 'batsman_runs' in descending order
# Drop duplicates to keep only the highest run-scorer for each season
# Finally, sort the results by 'season'
temp_df.groupby(['season', 'batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs', ascending=False).drop_duplicates(subset=['season'], keep='first').sort_values('season')
```
